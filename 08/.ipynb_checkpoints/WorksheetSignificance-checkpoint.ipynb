### HYPOTHESIS TESTING

```python
# test whether the mean murder is 8.5

murder_data = state_pd['Murder'].values
print(murder_data)
```

np.mean(murder_data)

```python
t_stat = (np.mean(murder_data) - 8.5) / (np.std(murder_data) / np.sqrt(len(murder_data)))
```

```python
df = len(murder_data)
```

```python
p_value = 2 * stats.t.sf(np.abs(t_stat), df)
```

```python
ci = stats.t.interval(0.95, df, loc=np.mean(murder_data), scale=np.std(murder_data)/np.sqrt(len(murder_data)))
```
